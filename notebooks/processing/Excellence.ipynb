# Academic Excellence

Process data from https://www.cde.ca.gov/ds/ad/assessmentdata.asp.

```python
%run notebooks/Setup.ipynb

import polars as polars
```

```python
caaspp = polars.read_csv("data/raw/dataquest/sb_ca2024_all_38_csv_v1.txt", separator='^')
# filter out null school names
caaspp = caaspp.filter(polars.col("School Name").is_not_null())
caaspp
```

```python
# filter for just claire lillienthal, all grades (grade 13), and all student groups (student group id 1)
caaspp_cl = caaspp.filter(
    (polars.col("School Name") == "Lilienthal (Claire) Elementary")
    & (polars.col("Grade") == 13)
    & (polars.col("Student Group ID") == 1)
)
caaspp_cl
```

```python
# filter by grade 13 and student group id 1 and test id 1 (ELA) 
caaspp_ela = caaspp.filter(
    (polars.col("Grade") == 13) & (polars.col("Student Group ID") == 1) & (polars.col("Test ID") == 1)
)

# keep only the % standard columns 
caaspp_ela = caaspp_ela.select(
    polars.col("School Name"),
    polars.col("Total Students Tested").cast(polars.Int32, strict=False),
    polars.col("Percentage Standard Exceeded").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Met").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Met and Above").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Nearly Met").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Not Met").cast(polars.Float64, strict=False)
)

# prefix numeric columns with "ELA"
caaspp_ela = caaspp_ela.rename({
    "School Name": "school_name",
    "Total Students Tested": "ELA Total Students Tested",
    "Percentage Standard Exceeded": "ELA Percentage Standard Exceeded",
    "Percentage Standard Met": "ELA Percentage Standard Met",
    "Percentage Standard Met and Above": "ELA Percentage Standard Met and Above",
    "Percentage Standard Nearly Met": "ELA Percentage Standard Nearly Met",
    "Percentage Standard Not Met": "ELA Percentage Standard Not Met"
})

caaspp_ela.write_csv(workspace_path.joinpath('data/processed/caaspp_ela.csv'))
caaspp_ela
```

```python
# filter by grade 13 and student group id 1 and test id 2 (math) 
caaspp_math = caaspp.filter(
    (polars.col("Grade") == 13) & (polars.col("Student Group ID") == 1) & (polars.col("Test ID") == 2)
)

# keep only the % standard columns 
caaspp_math = caaspp_math.select(
    polars.col("School Name"),
    polars.col("Total Students Tested").cast(polars.Int32, strict=False),
    polars.col("Percentage Standard Exceeded").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Met").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Met and Above").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Nearly Met").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Not Met").cast(polars.Float64, strict=False)
)

# prefix numeric columns with "ELA"
caaspp_math = caaspp_math.rename({
    "School Name": "school_name",
    "Total Students Tested": "Math Total Students Tested",
    "Percentage Standard Exceeded": "Math Percentage Standard Exceeded",
    "Percentage Standard Met": "Math Percentage Standard Met",
    "Percentage Standard Met and Above": "Math Percentage Standard Met and Above",
    "Percentage Standard Nearly Met": "Math Percentage Standard Nearly Met",
    "Percentage Standard Not Met": "Math Percentage Standard Not Met"
})

caaspp_math.write_csv(workspace_path.joinpath('data/processed/caaspp_math.csv'))
caaspp_math
```

```python
cast = polars.read_csv("data/raw/dataquest/cast_ca2024_all_38_csv_v1.txt", separator='^')

# filter out null school names
cast = cast.filter(
    polars.col("School Name").is_not_null()
    & (polars.col("Grade") == 13)
    & (polars.col("Student Group ID") == 1)
)

cast = cast.select(
    polars.col("School Name"),
    polars.col("Total Students Tested").cast(polars.Int32, strict=False),
    polars.col("Percentage Standard Exceeded").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Met").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Met and Above").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Nearly Met").cast(polars.Float64, strict=False),
    polars.col("Percentage Standard Not Met").cast(polars.Float64, strict=False)
)

cast = cast.rename({
    "School Name": "school_name",
    "Total Students Tested": "Science Total Students Tested",
    "Percentage Standard Exceeded": "Science Percentage Standard Exceeded",
    "Percentage Standard Met": "Science Percentage Standard Met",
    "Percentage Standard Met and Above": "Science Percentage Standard Met and Above",
    "Percentage Standard Nearly Met": "Science Percentage Standard Nearly Met",
    "Percentage Standard Not Met": "Science Percentage Standard Not Met"
})

cast.write_csv(workspace_path.joinpath('data/processed/cast.csv'))
cast
```
